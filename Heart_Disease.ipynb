```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# dataset
df = pd.read_csv('dataset.csv')
df.head()
```

```python
# describe
df.describe().T
```

```python
# correlation
corr=df.corr()
corr
```

```python
# heatmap
plt.figure(figsize=(15,10))
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
# isnull
df.isnull().sum().sum()
```

```python
# duplicate
df.duplicated().sum()
```

```python
# drop duplicate
df.drop_duplicates(inplace=True)
```

```python
# independent , dependent
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
# train test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
# predict
y_pred=lr.predict(x_test)
```

```python
# accuracy
from sklearn.metrics import accuracy_score
log_acc=accuracy_score(y_test,y_pred)*100
log_acc
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
Logistic_confusion=confusion_matrix(y_test,y_pred)
Logistic_confusion
```

```python
# heatmap
sns.heatmap(Logistic_confusion,annot=True,fmt='d')
```

```python
# classification report
from sklearn.metrics import classification_report
Logistic_class_report=classification_report(y_test,y_pred)
print(Logistic_class_report)
```

# Decision Tree

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
# predict
y_pred=dt.predict(x_test)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
confusion=confusion_matrix(y_test,y_pred)
confusion
```

```python
# accuracy
from sklearn.metrics import accuracy_score
dt_acc=accuracy_score(y_test,y_pred)*100
dt_acc
```

```python
# # confusion matrix
from sklearn.metrics import confusion_matrix
decision_confusion=confusion_matrix(y_test,y_pred)
decision_confusion
```

```python
# heatmap
sns.heatmap(decision_confusion,annot=True,fmt='d')
```

```python
# # classification report
from sklearn.metrics import classification_report
decision_class_report=classification_report(y_test,y_pred)
print(decision_class_report)
```

```python
#  Visualize the decision tree using the Graphviz package
from sklearn.tree import export_graphviz
import graphviz
dot_data=export_graphviz(dt,out_file=None,feature_names=x.columns,filled=True)
graph=graphviz.Source(dot_data,format='png')
graph
# graph.render("decision tree")  # Save the visualization as a PNG file
```

# Random Forest

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
# predict
y_pred=rf.predict(x_test)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
confusion=confusion_matrix(y_test,y_pred)
confusion
```

```python
# accuracy
rf_accuracy=accuracy_score(y_test,y_pred)*100
rf_accuracy
```

```python
#  Visualize the Random forest using the Graphviz package
from sklearn.tree import export_graphviz
import graphviz
dot_data=export_graphviz(rf.estimators_[0],out_file=None,feature_names=x.columns,filled=True)
graph=graphviz.Source(dot_data,format='png')
graph
```

```python
# # confusion matrix
from sklearn.metrics import confusion_matrix
random_forest_confusion=confusion_matrix(y_test,y_pred)
random_forest_confusion
```

```python
# # classification report
from sklearn.metrics import classification_report
rforest_class_report=classification_report(y_test,y_pred)
print(rforest_class_report)
```

```python
# accuracy comparation
print("Logistic Regression Accuracy:",log_acc)
print("Decision Tree Accuracy:",dt_acc)
print("Random Forest Accuracy:",rf_accuracy)
```
